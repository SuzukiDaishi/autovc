```python
import os
import pickle
import torch
import numpy as np
from math import ceil
import glob
```

```python
traindata = pickle.load(open('spmel/train.pkl', "rb"))
```

```python
out_data = []
for data in traindata:
    print(data[0], data[1].shape, np.load(glob.glob(f'spmel/{data[0]}/*.npy')[0]).shape)
    out_data.append([data[0], data[1], np.load(glob.glob(f'spmel/{data[0]}/*.npy')[0])])
with open('metadata.pkl', 'wb') as f:
    pickle.dump(out_data, f)
```
